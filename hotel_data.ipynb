# data analysis and visualizations

```python
# This code is a simple yet effective way to visualize the distribution of reservation cancellations, making it easier to understand the reservation status in your dataset.
# The bar chart clearly shows the count and percentage of canceled and non-canceled reservations.

cancelled_percentage = df["is_canceled"].value_counts(normalize=True)
print(cancelled_percentage)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not_Cancalled','Cancalled'],df["is_canceled"].value_counts(),edgecolor='k',width=0.9)
plt.show()
```

```python
# This code effectively visualizes the distribution of reservations among different hotels while highlighting the cancellation status.
# It provides a clear visual representation of how cancellations vary between different hotel types, making it easier to identify patterns and trends in the data.

plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels,_ =ax1.get_legend_handles_labels()
#ax1.legend(bbox_to_anchor(1,1))
plt.title('Reservation hotel different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not_cancalled','cancalled'])
plt.show()
```

```python
# This code allows you to analyze the cancellation behavior specifically for reservations at the resort hotel.
# It provides the percentage of reservations that were canceled and not canceled within this subset, helping you understand the cancellation patterns for this type of hotel.

resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# These operations are useful for analyzing the average daily rates ('adr') for reservations at both the resort hotel and the city hotel over time. 
# By grouping the data by reservation date and calculating the mean, you can track the trends and fluctuations in room rates for each type of hotel.
# This information can be valuable for understanding pricing strategies and making data-driven decisions related to hotel pricing.

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# The code effectively compares and visualizes the trends in average daily rates for both the resort hotel and the city hotel over time. 
# It provides a clear representation of how room rates have evolved for these two types of hotels.
# This information can be valuable for making pricing and revenue management decisions in the hotel industry

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 8))
plt.title('Average daily rate in city and resort hotel', fontsize=30)

# Convert pandas Series to numpy arrays before plotting
resort_hotel_index = resort_hotel.index.to_numpy()
resort_hotel_adr = resort_hotel['adr'].to_numpy()
city_hotel_index = city_hotel.index.to_numpy()
city_hotel_adr = city_hotel['adr'].to_numpy()

# Plot the data
plt.plot(resort_hotel_index, resort_hotel_adr, label='Resort Hotel')
plt.plot(city_hotel_index, city_hotel_adr, label='City Hotel')
plt.legend(fontsize=20)

# Add labels and other plot customizations if needed
plt.xlabel('Index', fontsize=16)
plt.ylabel('ADR', fontsize=16)

# Show the plot
plt.show()
```

```python
# The code effectively visualizes the distribution of canceled and not canceled reservations for each month. 
# It provides insight into the seasonal variations in reservation cancellations, helping you understand the monthly patterns in your data.

df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
plt.title('Reservation status per month',size=20)
plt.ylabel('number of observation',size=15)
plt.xlabel('month',size=15)
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
# The code effectively visualizes the total ADR per month for canceled reservations, providing insights into how ADR varies over different months. 
# This information can be valuable for analyzing revenue trends and making pricing decisions.

plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
# The code effectively visualizes the distribution of canceled reservations among the top 10 countries. 
# It provides a clear representation of which countries have the highest number of canceled reservations, allowing you to identify geographical trends in reservation cancellations.

canceled_data=df[df['is_canceled']==1]
top_10_country =canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('top_10_countries with reservation is canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
# The code effectively visualizes the trends in ADR for canceled and not canceled reservations over time. 
# It allows you to compare how ADR has varied for these two categories, providing insights into pricing and revenue management.

cancelled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_cancelled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20,8))
plt.title('Average daily rate' ,fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'].to_numpy(),not_cancelled_df_adr['adr'].to_numpy(),label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'].to_numpy(),cancelled_df_adr['adr'].to_numpy(),label='Cancelled')
plt.legend(fontsize=20)
plt.show()
```
